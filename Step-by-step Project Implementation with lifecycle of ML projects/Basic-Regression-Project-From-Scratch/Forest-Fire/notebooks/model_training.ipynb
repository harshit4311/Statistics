```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('/Users/harshit/Programming/Stats-Notes/Step-by-step Project Implementation with lifecycle of ML projects/Basic-Regression-Project-From-Scratch/Forest-Fire/dataset/Algerian_forest_fires_cleaned_dataset.csv')
```

```python
# drop month,day and yyear
df.drop(['day', 'month', 'year'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
# encoding
df['Classes']=np.where(df['Classes'].str.contains("not fire"), 0,1)
```

```python
df['Classes'].value_counts()
```

```python
# Independent And dependent features
X = df.drop('FWI', axis=1)
y = df['FWI']
```

```python
X.head()
```

```python
y.head()
```

```python
# Train-Test-Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train.shape, X_test.shape
```

```python
##Feature Selection based on correlation
X_train.corr()
```

```python
## Check for multicollinearity
plt.figure(figsize = (10,10))
corr = X_train.corr()
sns.heatmap(corr, annot=True)
```

```python
X_train.corr()
```

```python
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: 
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
# threshold -- Domain expertise
corr_features = correlation(X_train, 0.85)
```

```python
corr_features
```

```python
# drop features when correlation is more than 0.85 
X_train.drop(corr_features, axis=1, inplace=True)
X_test.drop(corr_features, axis=1, inplace=True)
X_train.shape, X_test.shape
```

## Feature Scaling or Standardisation

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

## Box plots to understand effects of Standard Scaler

```python
plt.subplots(figsize=(15, 5))
plt.subplot(1, 2, 1)

sns.boxplot(data = X_train)
plt.title('X_train Before Scaling')

plt.subplot(1, 2, 2)
sns.boxplot(data = X_train_scaled)
plt.title('X_train After Scaling')
```

## Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
linreg.fit(X_train_scaled, y_train)
y_pred = linreg.predict(X_test_scaled)
mae = mean_absolute_error(y_test,y_pred)
score = r2_score(y_test, y_pred)

print("Mean absolute error", mae)
print("R2 Score", score)
plt.scatter(y_test, y_pred)
```

## Lasso Regression Model
